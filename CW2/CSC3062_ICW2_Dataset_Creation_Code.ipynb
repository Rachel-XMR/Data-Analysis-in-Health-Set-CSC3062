A balanced dataset contains an equal number of samples in each group or category.
Aps: To create your balanced dataset, extract the last two digits of your student number. For instance, if your student number is Z1Z2Z3Z4Z5Z6Z7Z8, use Z7Z8, then sum these two digits (Z7 + Z8). The number of samples to select from each group in the original dataset will be determined as follows:
N = 134 – (Z7 + Z8)
For example, if the student number is 40268613, N = 134 – (1 + 3) = 134 – 4 = 130. Therefore, you will randomly choose 130 samples (without duplications) from each group in the original dataset.

```python
import pandas as pd # Importing pandas library as 'pd' for DataFrame operations
import numpy as np
```

```python
# Read original (ICW2) dataset downloaded from https://www.kaggle.com/datasets/rezarafiee/immune2009
df_ICW2 = pd.read_csv("ICW2_Dataset_CSC3062_2009_440_6ImmuneSubtypes_NoMissing_RR011119.csv", index_col=0)
```

```python
df_ICW2  # print the dataset 
```

```python
df_ICW2 = df_ICW2.T  # Transpose the original dataset
df_ICW2 # print the dataset
```

```python
# How many groups (or subgroup/class) do we have in this dataset?
np.unique(df_ICW2['Subgroup'])  # There are 6 Groups in the dataset
```

```python
df_ICW2['Subgroup'].value_counts()   # Check the total value for each group
```

### 1. Create a balanced dataset as explained in the next paragraph (Aps ). In a balanced dataset, we have the same number of samples in each group.

```python
# Assume that my Student number is 40268613, so the number of samples for each group should be 134 - (1 + 3) = 130
n1= 129 # for this case. This is just an example and you need to calclulate n1 based on your student number digits 
data1 = df_ICW2[df_ICW2['Subgroup'] == 'C1']  # Select samples with subgroup C1
data1 = data1.sample(n=n1)         # Random select n1 samples from class C1
data2 = df_ICW2[df_ICW2['Subgroup'] == 'C2']  # Select samples with subgroup C2
data2 = data2.sample(n=n1)         # Random select n1 samples from class C2
data3 = df_ICW2[df_ICW2['Subgroup'] == 'C3']  # Select samples with subgroup C3
data3 = data3.sample(n=n1)         # Random select n1 samples from class C3
data4 = df_ICW2[df_ICW2['Subgroup'] == 'C4']  # Select samples with subgroup C4
data4 = data4.sample(n=n1)         # Random select n1 samples from class C4
data5 = df_ICW2[df_ICW2['Subgroup'] == 'C5']  # Select samples with subgroup C5
data5 = data5.sample(n=n1)         # Random select n1 samples from class C5
data6 = df_ICW2[df_ICW2['Subgroup'] == 'C6']  # Select samples with subgroup C6
data6 = data6.sample(n=n1)         # Random select n1 samples from class C6
frame = [data1, data2, data3, data4, data5, data6]
balenced_df_ICW2 = pd.concat(frame)
# Check sample size for each group
print(balenced_df_ICW2['Subgroup'].value_counts())
balenced_df_ICW2
```
